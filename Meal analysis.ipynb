```python
train=pd.read_csv('C:/Users/hp/Desktop/Python Courses/Meal project/train.csv')
train
```

```python
test=pd.read_csv('C:/Users/hp/Desktop/Python Courses/Meal project/test.csv')
test
```

```python
test['num_orders']='0'
test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train=train.drop(['id', 'num_orders'], axis=1)
y_train=train['num_orders']
```

```python
x_test=test.drop(['id','num_orders'], axis=1)
y_test=test['num_orders']
```

```python
regr=LinearRegression()
```

```python
regr.fit(x_train,y_train)
```

```python
regr.predict(x_test)
```

```python
pred=regr.predict(x_test)
```

```python
regr.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,pred)
```

```python
import math
rmse=math.sqrt(mse)
100*rmse
```

```python
pred=pd.DataFrame(pred)
pred=pred.rename(columns={0:'num_orders'})
```

```python
test.update(pred, overwrite=True)
test
```

```python
test.to_csv('result.csv')
```
